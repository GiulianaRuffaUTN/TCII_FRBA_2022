1. Antes que nada realizamos el análisis de tensiones y corrientes del circuito.

A partir de esto, podemos establecer las siguientes relaciones:

$ {{V_{1} - V_2}\over R_{1}+R_{2}} = {{V_{1} - V^{-}}\over R_{1}} = {V^{-} - {V_{2}}\over R_{2}} = I_1 $

$ {V_{1}\over R_{3}+{1 \over CS}} = I_{2} $

$V^{+}=I_2 . R_3$

$V^{-}= V_1 - I_1 . R_1$

Luego: $\; V_{2} = V^{-} - I_1 . R_{2} $

Además: $\;\; V^{+} = V^{-}:$

$I_2 . R_3 = V_1 - I_1 . R_1$ 

$ \rightarrow \; {I_1 = {{V_1 - I_2 . R_3}\over  R_1}}$

$ \rightarrow \; {I_2 = {{V_1 - I_1 . R_1}\over  R_3}}$

Finalmente, planteamos la ecuación de $V_2$ y reemplazamos con las expresiones obtenidas: 

$ V_{2} = V^{-} - I_1 . R_{2} = (V_1 - I_1 . R_1) - I_1 . R_{2} = V_1 - I_1 . (R_1 - R_{2}) $

$= V_1 - ({{V_1 - I_2 . R_3}\over  R_1}) . (R_1 - R_{2}) = V_1 - ({{V_1 - {V_{1}\over R_{3}+{1 \over CS}} . R_3}\over  R_1}) . (R_1 - R_{2}) $

A partir de esto, despejamos la transferencia de módulo ${V_2 \over V_1}$:

${V_2 \over V_1} (S)= {{SCR_3 - {R_2 \over R_1}} \over {S CR_3 + 1}} = {{S - {R_2 \over R_1CR_3}} \over {S + {1\over CR_3}}}$

Luego, obtenemos la expresión de módulo, fase y el diagrama de polos y ceros:

• Módulo:

${V_2 \over V_1} (jw)={{w^{2} + ({R_2 \over R_1CR_3})^{2}} \over {w^{2} + ({1\over CR_3})^{2}}}$

• Fase:

${\Phi} (w)= tg^{-1} ({w \over {{R_2 \over R_1 }.{1 \over {C.R_3}}}}) - tg^{-1} ({w \over {{1 \over {C.R_3}}}})$
